## Hitung delay penerbangan (in seconds)

# Hitung jumlah penerbangan

```python
import pandas as pd
```

```python
# File path
file_path = '/content/drive/MyDrive/Colab Notebooks/Skripsi/dataset/2020/sorted_data2020.csv'

# Membaca file CSV
df = pd.read_csv(file_path)
df
```

```python
# Pastikan kolom tanggal ada, sesuaikan namanya (contoh: 'Tanggal')
# Konversi kolom tanggal ke format datetime
df['TGL'] = pd.to_datetime(df['TGL'])

# Tambahkan kolom Bulan dan Tahun
df['bulan'] = df['TGL'].dt.month
df['tahun'] = df['TGL'].dt.year

# Mengelompokkan data berdasarkan Tahun dan Bulan, lalu menghitung jumlah record
summary = df.groupby(['tahun', 'bulan']).size().reset_index(name='Jumlah_Penerbangan')

# Tampilkan hasil
print(summary)
```

# Visualisasi Plot Trafik Delay Penerbangan

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = '/content/drive/MyDrive/Colab Notebooks/Skripsi/dataset/normalizedFixedSorted_data_2020-2024.csv'

# Membaca file CSV
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Pastikan kolom tanggal ada, sesuaikan namanya (contoh: 'Tanggal', 'SCH', 'ACT')
df['TGL'] = pd.to_datetime(df['TGL'], errors='coerce')

# Tambahkan kolom Bulan dan Tahun
df['Bulan'] = df['TGL'].dt.month
df['Tahun'] = df['TGL'].dt.year
```

```python
# Agregasi jumlah delay per bulan dan tahun
summary_delay = df.groupby(['Tahun', 'Bulan'])['delay'].sum().reset_index(name='Total_Delay')

# Membuat plot deret waktu
plt.figure(figsize=(10, 6))

# Membuat label waktu dalam format "Tahun-Bulan"
summary_delay['Tahun-Bulan'] = summary_delay['Tahun'].astype(str) + '-' + summary_delay['Bulan'].astype(str)

# Plot total delay per bulan dan tahun
plt.plot(summary_delay['Tahun-Bulan'], summary_delay['Total_Delay'], marker='o', color='b', linestyle='-', linewidth=2, markersize=5)

# Menambahkan label dan judul
plt.xlabel('Tahun-Bulan')
plt.ylabel('Total Delay (menit)')
plt.title('Trafik Delay Penerbangan Domestik per Bulan dan Tahun')
plt.xticks(rotation=45)  # Agar label lebih mudah dibaca
plt.grid(True)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Agregasi jumlah delay per bulan dan tahun
summary_delay = df.groupby(['Tahun', 'Bulan'])['delay'].sum().reset_index(name='Total_Delay')

# Membuat label waktu dalam format "Tahun-Bulan" yang lebih sederhana
summary_delay['Tahun-Bulan'] = summary_delay['Tahun'].astype(str) + '-' + summary_delay['Bulan'].astype(str).str.zfill(2)

# Set index ke kolom 'Tahun-Bulan'
summary_delay.set_index('Tahun-Bulan', inplace=True)

# Membuat plot deret waktu dengan style mirip dengan contoh
plt.figure(figsize=[12, 8])  # Set dimensi figure lebih besar
plt.plot(summary_delay.index, summary_delay['Total_Delay'], marker='o', color='g', linestyle='-', linewidth=2, markersize=6)  # Plot dengan warna hijau

# Menambahkan judul, label, dan pengaturan lainnya
plt.title('Trafik Total Delay Penerbangan Domestik per Bulan', fontsize=16)
plt.ylabel('Total Delay (menit)', fontsize=14)
plt.xlabel('Tahun-Bulan', fontsize=14)

# Menetapkan interval setiap 4 bulan
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(ticks=range(0, len(summary_delay.index), 4), labels=[month_labels[i % 12] for i in range(0, len(summary_delay.index), 4)], rotation=45)

plt.grid(True)  # Menampilkan grid

# Menampilkan plot
plt.tight_layout()
plt.show()
```
